```python
import settings

import joblib

payload = joblib.load(str(settings.MODEL_HOME / "svm/svm_emotion_model.pkl"))
best_model = payload["model"]
vectorizer = payload["vectorizer"]

print("Successfully loaded Model & Vectorizer.")
```

```python
def predict_emotion(text: str, model=best_model, vectorizer=vectorizer):
    clean_text = text.strip()
    X = vectorizer.transform([clean_text])
    pred_label = model.predict(X)[0]
    return pred_label
```

```python
while True:
    s = input("Input your sentence (q to quit) : ").strip()
    if s.lower() == "q":
        print("Quit")
        break
    if not s:
        continue

    label = predict_emotion(s)
    print(f"Predicted emotion label: {label}")
    print("=" * 50)
```
